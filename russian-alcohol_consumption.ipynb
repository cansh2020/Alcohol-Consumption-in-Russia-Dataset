# 2. Data Understanding
The dataset I shall use is from https://github.com/datacamp/careerhub-data/commit/33df87ef27afbcb8ab6f46e5f0c4ebba76c9791a#diff-94d8fe79d4129fd908cd13dfcca7aa269526c796645b3bce118d96e8f8967c86

It contains the values of the consumed alcohol volumes of beer, brandy champagne, vodka and wine for different regions in Russia during the period between 1998 and 2016.

Dataset has 1615 rows and 7 columns. Keys for columns:

"year" - year (1998-2016)

"region" - name of a federal subject of Russia. It could be oblast, republic, krai, autonomous okrug, federal city and a single autonomous oblast

"wine" - sale of wine in litres by year per capita

"beer" - sale of beer in litres by year per capita

"vodka" - sale of vodka in litres by year per capita

"champagne" - sale of champagne in litres by year per capita

"brandy" - sale of brandy in litres by year per capita

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('russia_alcohol.csv')
df.head()
```

# 3. Prepare Data

```python
a=df.iloc[:,1:8]
```

```python
pd.set_option('display.max_rows', 100)
row_has_NaN = a.isnull().any(axis=1)
rows_with_NaN = a[row_has_NaN]
rows_with_NaN
```

```python
a=a.fillna(0)
```

NaN values are replaced by 0

# 4. Modeling

```python
alcohol = ["beer","vodka","wine","champagne","brandy"]
a["total_volume"] = a[alcohol].sum(axis=1)
a["mean"] = a[alcohol].mean(axis=1)
a["standard_deviation"] = a[alcohol].std(axis=1)
a.head()
```

```python
a=a.iloc[:,1:11]
a.head()
```

```python
a_data = a.groupby(['region'], as_index=True).sum()
print(a_data)
```

# 5. Analysis

```python
print(a_data.sort_values(by="mean",ascending=False))
```

```python
print(a_data.sort_values(by="standard_deviation",ascending=False))
```

The tables are sorted by mean value and standard daviation value now our motive is to select top 10 common values(except saint Petersburg) among databases that can be done manually we will be taking mean values and standard daviation value as a criteria to select regions for our campaign.
